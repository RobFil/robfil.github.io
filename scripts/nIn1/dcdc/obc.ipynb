## On-Board Charger (OBC) – Functional Block Structure and Operating Modes

An automotive OBC is a **multi-mode power conversion system**.  
Depending on charging mode and system features, different functional blocks are active, bypassed, or reused.

---

## 1. Grid- / Source-Side Power Conversion

- **EMI Filter & Protection**  
  Grid and source protection, EMC compliance  
  *(mandatory for AC input, reduced for DC input)*

- **AC/DC Converter with PFC**  
  AC → DC conversion with unity power factor  
  *(used only during AC charging)*

- **High-Voltage DC Link**  
  Regulated intermediate DC voltage  
  - ~380–420 V for 400 V vehicle architectures  
  - ~750–850 V for 800 V vehicle architectures  
  Energy buffering and decoupling between stages

---

## 2. Isolated DC/DC Converter (Single Functional Unit)

The isolated DC/DC converter is a single functional unit consisting of three inseparable sub-blocks:

- **DC/AC Inverter**  
  Converts DC link voltage into high-frequency AC  
  (half-bridge or full-bridge)

- **High-Frequency Transformer**  
  Provides galvanic isolation and voltage adaptation

- **AC/DC Rectifier**  
  Converts secondary-side AC back to DC  
  (diodes or synchronous rectification)

> This DC/DC block may operate in **unidirectional** or **bidirectional** mode  
> depending on topology (LLC vs. DAB) and enabled features (e.g. V2G).

---

## 3. Battery-Side Interface

- **DC Output Filter**  
  Ripple reduction and current smoothing

- **Battery Interface & Contactors**  
  Safe connection to the HV battery  
  Isolation monitoring and current sensing

---

## 4. Control & Supervision (Cross-Functional)

- Digital controller (MCU / DSP)
- PFC current control (AC charging)
- DC/DC control  
  - LLC: frequency modulation  
  - DAB: phase-shift modulation
- Charging strategy (CC / CV)
- Diagnostics, safety, and vehicle communication

---

## Supported Charging and Energy Flow Modes

### AC Charging – Single Phase
- Input: 230 V AC
- Active blocks:
  - EMI filter
  - AC/DC with PFC
  - DC link (~400 V)
  - Isolated DC/DC
- Typical power: 3.3–7.4 kW

---

### AC Charging – Three Phase
- Input: 400 V AC (3-phase)
- Active blocks:
  - EMI filter
  - 3-phase AC/DC with PFC
  - DC link (~400 V or ~800 V)
  - Isolated DC/DC
- Typical power: 11–22 kW

---

### DC Charging – 400 V Architecture
- Input: external DC charger (~200–500 V DC)
- Active blocks:
  - Protection & contactors
  - Optional isolated DC/DC (depending on OEM strategy)
- AC/DC with PFC is bypassed

---

### DC Charging – 800 V Architecture
- Input: external DC charger (~400–1000 V DC)
- Active blocks:
  - Protection & contactors
  - Often an isolated bidirectional DC/DC (DAB)
- Enables battery voltage matching and current control

---

### Vehicle-to-Grid (V2G / V2H / V2L)
- Bidirectional energy flow
- Required features:
  - **Bidirectional isolated DC/DC (typically DAB)**
  - Active rectification on both sides
  - Precise power and phase control
- AC/DC stage operates as inverter during export

> V2G fundamentally changes the OBC from a charger into a **grid-connected energy interface**.

---

## Topology Impact of Trend Features

| Feature | Required DC/DC Topology |
|------|-------------------------|
| AC charging only | LLC sufficient |
| DC charging (unidirectional) | LLC or DAB |
| Wide voltage range (400 V ↔ 800 V) | DAB preferred |
| Bidirectional power flow | **DAB mandatory** |
| V2G / V2H / V2L | **DAB mandatory** |

---

## Key Takeaways

- The OBC is a **multi-mode power conversion platform**
- AC/DC with PFC is only required for **AC charging**
- DC charging may bypass internal power stages
- **V2G and bidirectional charging fundamentally require DAB**
- Modern designs increasingly merge OBC and DC/DC into a **single bidirectional power unit**

> Trend features do not add blocks —  
> they **change how existing blocks are operated and controlled**.


## LLC vs. DAB – Topology Comparison for Automotive OBC Applications

LLC and DAB are both isolated DC/DC topologies, but they are optimized for **fundamentally different system goals**.
The choice is driven by charging mode, voltage range, power flow direction, and control requirements.

---

## 1. Fundamental Operating Principle

### LLC (Resonant Converter)
- Resonant power transfer (Lr–Cr–Lm)
- Fixed duty cycle
- Output regulated by **switching frequency**
- Soft switching achieved by resonance physics

### DAB (Dual Active Bridge)
- Two actively switched bridges
- Power transfer via **phase shift**
- Fixed switching frequency
- Soft switching achieved by controlled current direction

---

## 2. Control Philosophy

| Aspect | LLC | DAB |
|-----|-----|-----|
| Main control variable | Switching frequency | Phase shift |
| Control complexity | Low | Medium–High |
| Software dependency | Low | High |
| Physics vs. software | Physics-dominated | Software-dominated |

---

## 3. Voltage Range Capability

| Criterion | LLC | DAB |
|--------|-----|-----|
| Narrow voltage range | Excellent | Good |
| Wide voltage range | Poor | Excellent |
| 400 V ↔ 800 V systems | Not suitable | Ideal |

LLC requires the operating point to remain close to resonance.  
DAB tolerates large voltage ratios without loss of controllability.

---

## 4. Power Flow Direction

| Mode | LLC | DAB |
|----|-----|-----|
| Unidirectional | Yes | Yes |
| Bidirectional | No (impractical) | **Yes (native)** |
| V2G / V2H / V2L | No | **Mandatory** |

Bidirectional power flow is **structural** in DAB and **structurally excluded** in LLC.

---

## 5. Charging Mode Compatibility

| Charging Mode | LLC | DAB |
|-------------|-----|-----|
| AC charging (single phase) | ✔️ | ✔️ |
| AC charging (three phase) | ✔️ | ✔️ |
| DC charging (400 V) | Limited | ✔️ |
| DC charging (800 V) | ❌ | ✔️ |
| V2G / bidirectional | ❌ | ✔️ |

---

## 6. Efficiency Characteristics

### LLC
- Peak efficiency at nominal point (>97%)
- Efficiency drops quickly off-design
- Excellent for **fixed operating conditions**

### DAB
- Slightly lower peak efficiency
- High efficiency over wide operating range
- Predictable losses

---

## 7. Current Stress and Scaling

| Aspect | LLC | DAB |
|------|-----|-----|
| Circulating current | Load-dependent | Control-defined |
| Light-load behavior | Poor | Excellent |
| High-power scalability | Limited | Excellent |

DAB allows **explicit control of current**, LLC does not.

---

## 8. Magnetics and Hardware Complexity

| Aspect | LLC | DAB |
|------|-----|-----|
| Resonant tank | Required | Not required |
| Transformer utilization | Load-dependent | Predictable |
| Component count | Lower | Higher |

LLC trades flexibility for simplicity.

---

## 9. Typical Automotive Use Cases

### LLC – Best Fit
- AC charging only
- Fixed battery voltage (≈ 400 V)
- No bidirectional power flow
- Highest efficiency at nominal point

### DAB – Best Fit
- AC + DC charging
- 400 V / 800 V platforms
- Bidirectional power flow (V2G)
- Unified OBC + DC/DC architectures

---

## 10. Strategic Industry Trend

| Generation | Dominant Topology |
|----------|-------------------|
| Early EVs | LLC |
| Current platforms | LLC + DAB |
| Next generation | **DAB-centric** |

---

## Key Takeaways

- **LLC is a high-efficiency specialist**
- **DAB is a flexible energy router**
- Trend features do not add blocks — they **change topology choice**
- V2G and 800 V architectures make **DAB unavoidable**

> LLC optimizes efficiency at one point.  
> DAB optimizes controllability over all points.
