# MSRFT Harmonic Mapping and Compensation

The animations show how space-vector harmonics transform between reference frames and how they are extracted and compensated using Multi Synchronous Reference Frame Transformation.

In the stator frame (αβ):
- Fundamental rotates at +ω  
- 5th harmonic rotates at −5ω (negative sequence)  
- 7th harmonic rotates at +7ω (positive sequence)

The dq frame rotates at +ω, therefore all components are observed with the relative speed

ω_dq = ω_signal − ω

which gives:
- 5th: (−5ω) − (+ω) = −6ω  
- 7th: (+7ω) − (+ω) = +6ω  

Thus both appear as 6th-order components in dq, only with opposite rotation direction.

The first two animations visualize this frequency shift from αβ to dq.

The third script and its plots shows the MSRFT principle:
additional synchronous frames rotating at ±6ω are introduced so that
- the 5th harmonic becomes DC in the −6ω frame,
- the 7th harmonic becomes DC in the +6ω frame.

These DC components are extracted and directly used to generate counteracting signals.
After inverse transformations back to dq and abc, the compensation currents cancel the original harmonics.

This demonstrates that MSRFT converts oscillatory harmonic disturbances into DC quantities,
which can be measured, regulated, and eliminated using standard control techniques.
